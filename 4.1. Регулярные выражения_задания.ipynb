```python
import re
text = "Nice!Good job#Well done- Perfect"
split = re.split(r"[!#-]", text)

split
```

üîπ **–ó–∞–¥–∞–Ω–∏–µ 2. –ó–∞–º–∞—Å–∫–∏—Ä–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –ø–æ—Å–ª–µ–¥–Ω–∏–µ —á–µ—Ç—ã—Ä–µ —Ü–∏—Ñ—Ä—ã –Ω–æ–º–µ—Ä–∞ –∫–∞—Ä—Ç—ã –Ω–∞ –∑–≤–µ–∑–¥–æ—á–∫–∏**

```python
text = "–ù–æ–º–µ—Ä –∫–∞—Ä—Ç—ã: 1234 5678 9012 3456"
sub = re.sub(r"\d{4}$", r"****", text)

sub
```

üîπ **–ó–∞–¥–∞–Ω–∏–µ 3. –í—ã–≤–µ—Å—Ç–∏ —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—É—é —Ñ–∞–º–∏–ª–∏—é —Å –æ–∫–æ–Ω—á–∞–Ω–∏–µ–º "–∏–π"**

```python
text = "–í —Å–ø–∏—Å–∫–µ —Ñ–∞–º–∏–ª–∏–π –±—ã–ª–∏ –ò–≤–∞–Ω–æ–≤, –Ø–Ω–æ–≤—Å–∫–∏–π –∏ –í–∞—Å–∏–ª—å–µ–≤"
search = re.search(r"[–ê-–Ø–Å][–∞-—è—ë]+–∏–π", text)

search.group(0)
```

üîπ **–ó–∞–¥–∞–Ω–∏–µ 4. –í—ã–≤–µ—Å—Ç–∏ –≤–µ—Å—å —Ç–µ–∫—Å—Ç –º–µ–∂–¥—É –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è–º–∏, —Ç–æ –µ—Å—Ç—å –º–µ–∂–¥—É –∑–≤–µ–∑–¥–æ—á–∫–∞–º–∏**

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –Ω–µ –∑–∞–±—ã—Ç—å, —á—Ç–æ –Ω—É–∂–Ω–æ —ç–∫—Ä–∞–Ω–∏—Ä–æ–≤–∞—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Å–∏–º–æ–ª–≤—ã (–æ–±—Ä–∞—Ç–Ω—ã–π —Å–ª—ç—à)

```python
text = "–∫–æ–¥ * –∏—Å–ø—Ä–∞–≤–∏—Ç—å –≤–æ—Ç –∑–¥–µ—Å—å! –¥–æ–±–∞–≤–∏—Ç—å –ø—Ä–æ–≤–µ—Ä–∫—É * –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ"
search = re.search(r"\*(.*?)\*", text)
print(search.group(1))
```

üîπ **–ó–∞–¥–∞–Ω–∏–µ 5. –í—ã–≤–µ—Å—Ç–∏ –≤—Å–µ —Ü–µ–Ω—ã –∏–∑ —Ç–µ–∫—Å—Ç–∞**

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: —Ü–µ–Ω–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏–∑ –ª—é–±–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä –∏ –º–æ–∂–µ—Ç –±—ã—Ç—å —É–∫–∞–∑–∞–Ω–∞ —Å –∫–æ–ø–µ–π–∫–∞–º–∏, —Ç–æ –µ—Å—Ç—å –æ—Ç 1 –¥–æ 2 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ —Ç–æ—á–∫–∏

```python
text = "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ: 5 —à—Ç—É–∫. –¶–µ–Ω–∞: 199.90 —Ä—É–±–ª–µ–π. –ó–∞–≤—Ç—Ä–∞ –±—É–¥–µ—Ç —Ü–µ–Ω–∞: 220 —Ä—É–±–ª–µ–π"
findall = re.findall(r"\–¶–µ–Ω–∞:\s*(\d+(?:\.\d{1,2})?)", text, flags=re.IGNORECASE)
findall
```
